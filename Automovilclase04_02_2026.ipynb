```python
class Automovil: #clase
  def __init__(self, marca , modelo , color): #Constructor
    self.marca = marca  #atributos
    self.modelo = modelo
    self.color = color
  def mostrar_info(self):  #metodos
    print(f"La marca de el vehiculo: {self.marca}")
    print(f"El modelo del vehiculo: {self.modelo}")
    print(f"El color del vehiculo: {self.color}")

def main():
    marca = input("Ingrese la marca del vehiculo: ")
    modelo = input("Ingrese el modelo del vehiculo: ")
    color = input("Ingrese el color del vehiculo: ")
    vehiculo1 = Automovil(marca , modelo , color) # utilizo el registro de los atributos  uno a uno
    vehiculo2 = Automovil("Mercedez benz", 2026, "Palo rosa")# Ingreso los datos directamente
    vehiculo1.mostrar_info() #objeto 1
    vehiculo2.mostrar_info() #objeto 2
if __name__ == "__main__":
      main()
```
