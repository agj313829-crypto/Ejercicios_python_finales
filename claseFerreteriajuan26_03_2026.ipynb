```python
 #inventario de una ferreteria
 #metodos append() agregar datos
 #simular un sistema de control de inventarios
 # 1. Agregar nuevos productos append()
 # 2. Insertar productos en una posicion especifica insert()
 # 3. Eliminar productos agotados remove() o pop()
 # Ordenar productos por nombre y por precio sort()
 # Invertir el orden de la lista en su presentacion reverse()
 # Validar las entradas de los datos
def mostrar_inventario(productos):
   print("Inventario actual")
   if not productos:
    print("Inventario vacio, no hay productos")
    return
   for i, producto in  enumerate(productos, start=1):
    print(f"Ele-{i}. {producto['nombre']} - ${producto['precio']:.2f}")
    print(f"${producto['precio']:.2f} -> Ele-{i}. {producto['nombre']}")

#1. Agregar nuevos productos append()
def agregar_producto(productos, nombre,precio):
  #validar entrada de datos
  if isinstance(nombre, str) and nombre.strip() and nombre != "" and isinstance(precio, (int, float)) and precio > 0:
    productos.append({"nombre": nombre.title().strip(),"precio": precio })
    print(f"El producto '{nombre.title()}' fue agregado exitosamente..")
  else:
    print("Datos invalidos, nombre debe ser un texto y el precio un valor positivo")




def main():
    inventario = [
     {"nombre": "Taladro", "precio": 150000},
     {"nombre": "Martillo ", "precio": 40000},
     {"nombre": "Destornillador ", "precio": 150000},
    ]
    mostrar_inventario(inventario)
    agregar_producto(inventario, "Tornillo", 200)
    mostrar_inventario(inventario)


if __name__ == "__main__":
  main()
```
