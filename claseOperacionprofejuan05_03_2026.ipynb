```python
class Operacion:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def validar(self):
        if not isinstance(self.a, (int,float)) or not isinstance(self.b, (int,float)):
            raise ValueError("Los valores deben ser numericos")
    def mostrar_datos(self):
        print(f"Los datos proporcionados son: {self.a} y {self.b}")
    def calcular(self):
        print("**REALIZANDO LA OPERACION MATEMATICA**")
class Suma(Operacion):
      def calcular(self):
          super().validar()
          super().mostrar_datos()
          super().calcular()
          resultado = self.a + self.b
          print(f"El resultado de la SUMA es: {resultado}")

class Resta(Operacion):
      def calcular(self):
        super().validar()
        super().mostrar_datos()
        super().calcular()
        resultado = self.a - self.b
        print(f"El  resultado de la RESTA es: {resultado}")


class Multiplicacion(Operacion):
      def calcular(self):
        super().validar()
        super().mostrar_datos()
        super().calcular()
        resultado = self.a * self.b
        print(f"El resultado de la MULTIPLICACION es: {resultado}")



class Division(Operacion):
      def calcular(self):
        super().validar()
        super().mostrar_datos()
        super().calcular()
        if self.b == 0:
          print("No se puede dividir entre 0")
        else:
          resultado = self.a / self.b
        print(f"El resultado de la DIVISION es: {resultado}")



class Potencia(Operacion):
  def calcular(self):
    super().validar()
    super().mostrar_datos()
    super().calcular()
    if self.b >3:
      print("La potencia solo se puede hasta 3")
    else:
      resultado = self.a ** self.b
      print(f"El resultado de la potencia es: {resultado}")


def main():

    num1 = int(input("Ingrese el primer numero"))
    num2 = int(input("Ingrese el segundo numero"))
    print("OPERACION SUMA")
    operacion1 = Suma(num1,num2)
    operacion1.calcular()



    num3 = int(input("Ingrese el primer numero"))
    num4 = int(input("Ingrese el segundo numero"))
    print("OPERACION RESTA")
    operacion2 = Resta(num3,num4)
    operacion2.calcular()

    num5 = int(input("Ingrese el primer numero"))
    num6 = int(input("Ingrese el segundo numero"))
    print("OPERACION MULTIPLICACION")
    operacion3 = Multiplicacion(num5,num6)
    operacion3.calcular()


    num7 = int(input("Ingresa el primer numero"))
    num8 = int(input("Ingrese el segundo numero"))
    print("OPERACION DIVISION")
    operacion4 = Division(num7,num8)
    operacion4.calcular()

    num9 = int(input("Ingrese el primer numero"))
    num10 = int(input("Ingrese el segundo numero"))
    print("OPERACION POTENCIA")
    operacion5 = Potencia(num9,num10)
    operacion5.calcular()





if __name__ == "__main__":
  main()






6
```
