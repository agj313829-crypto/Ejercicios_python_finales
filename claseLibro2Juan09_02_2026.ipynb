```

class Libro:   #clase
  def __init__(self,titulo, precio):   #constructor
    self.__titulo = titulo   #atributos
    self.__precio = precio
  def get_precio(self):    #metodo , devolver el valor contenido en el atributo precio
    return self.__precio
  def set_precio(self, nuevo_precio):  #metodo set para modificar el precio , o actualizarlo
    if isinstance(nuevo_precio, (int, float)) and nuevo_precio >0:
      self.__precio = nuevo_precio
    else:
      print("El precio debe ser un numero positivo.")
  def mostrar_informacion(self):
    print(f"Titulo: {self.__titulo}")
    print(f"Precio: {self.__precio:,.2f}")
def main():
    print("\nSISTEMA DE LIBROS\n")
    libro1 = Libro("Python ADSO", 120000)
    libro2 = Libro("Java ADSO" , 100000)
    print("Los datos de los libros son: ")
    print("Los datos del libro 1 son: \n")
    libro1.mostrar_informacion()
    #vamos a mostrar el precio actual del libro 1
    print("\nEl precio actual del libro 1 es:" , libro1.get_precio())
    # vamos a modificar el precio del libro 1
    print("modificacion del precio del libro 1")
    libro1.set_precio(-150000)
    libro1.set_precio(150000)
    #vamos a mostrar el nuevo precio actual de el libro 1
    print("\nEl precio de el nuevo libro  1 es" ,libro1.get_precio())
    print("Los datos nuevo del libro 1 son: \n")
   #libro1.mostrar_informacion()
    print("\n FIN DEL PROGRAMA LIBRO  :)")
if __name__ == "__main__":
    main()
```
