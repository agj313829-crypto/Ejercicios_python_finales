```python
from abc import ABC, abstractmethod

# crearemos la clase abstracta , clase padre
class Figura(ABC):
  @abstractmethod
  def calcular_area(self):
    # este metodo se utlizara en todas las subclases de forma OBLIGATORIA
    pass
# creamos clases hijas de la figura
class Circulo(Figura):
  def __init__(self, radio):
    self.radio = radio
  def calcular_area(self):
     print("***** CALCULANDO AREA *******\n")
     return (3.1416 * (self.radio ** 2))

class Triangulo(Figura):
  def __init__(self, base , altura):
    self.base = base
    self.altura = altura
  def calcular_area(self):
    print("***** CALCULANDO AREA *******\n")
    return ( 3 * (self.base * self.altura)) / 2



class Rectangulo(Figura):
  def __init__(self , base , altura):
    self.base = base
    self.altura = altura
  def calcular_area(self):
    print("**** CALCULANDO AREA ***** \n")
    return( (self.base * self.altura))



class Cubo(Figura):
  def __init__(self ,lado):
    self.lado = lado
  def calcular_area(self):
    print("**** CALCULANDO AREA ****\n")
    return ( self.lado ** 2 )



class Piramide(Figura):
  def __init__(self, base, altura):
    self.base = base
    self.altura = altura
  def calcular_area(self):
    print("**** CALCULANDO AREA ***** \n")
    return (self.base * self.altura)

    #def calcular_area(self):
    pass

def mostrar_area(figura: Figura):
  # Este metodo recibe cualquier objeto que sea de cualquier tipo figuta
  # ejecuta el metodo calcular:area() , sin interesar de que tipo de hijo es .
  print(f"El Area: {figura.calcular_area():.2f}")

def main():
  circulo1 = Circulo(8)
  mostrar_area(circulo1)

  triangulo2 = Triangulo(7,9)
  mostrar_area(triangulo2)

  rectangulo3 = Rectangulo(8,5)
  mostrar_area(rectangulo3)

  cubo4 = Cubo(3)
  mostrar_area(cubo4)

  piramide5 = Piramide(6,7)
  mostrar_area(piramide5)



if __name__ == "__main__":
   main()
```

```python
from abc import ABC , abstractmethod
# clase con encapsulamiento
class Empleado(ABC):
  def __init__(self, nombre: str) -> None:
  # crearemos  nombre como privado
   self._nombre = None
   self.nombre = nombre # llamar al setter automaticamente
   # CREAREMOS EL GETTER
   @property
   def nombre(self) -> str:
     # tendre un acceso seguro al nombre del empleado
     return self._nombre

     #crearemos el SETTER
   @nombre.setter
   def nombre(self, valor: str) -> None:
      if isinstance(valor , str) and valor.strip():   # sea texto y no tenga la instacia en blanco .
        self._nombre = valor.strip() # asigna el valor adicional quita los aspectos en blanco del inicio & fin de la cadena o palabra
      else:
        raise ValueError("El nombre debe ser un texto valido")
  @abstractmethod
  def trabajar(self) -> None:
    pass
    #ahora crearemos a los hijos

class Gerente(Empleado):
    def trabajar(self) -> None:
      print(f"{self.nombre} Esta realizando la administracion de la empresa")
class Desarrollador(Empleado):
    def trabajar(self) -> None:
       print(f"{self.nombr}  Esta programando una solucion en python")

    def ejecutar_tarea(empleado: Empleado) -> None:
      empleado.trabajar()
    def main():
      print("******** INICIO DEL PROGRAMA DE EMPLEADOS EMPRESA SENA *******")
      desarrollador1 = Desarrollador("Joseph Martinez")
      gerente1 = Gerente("Pedro Moreno")
      print("polimosfismo con GETTER ")
      ejecutar_tarea(desarrollador1)
      ejecutar_tarea(gerente1)
      print("polimorfismo con SETTER em clase ABSTRACTA")


    if __name__ == "__main__":
      main()


```
