```python
from abc import ABC, abstractmethod
class GenerarReporte(ABC):
  @abstractmethod
  def generar_reporte(self) ->None:
    # Este metodo lo usaremos obligatoriamente en todas las clases hijas
     pass


class SistemaPago(ABC):
    def __init__(self, monto : float) ->None:
      self.monto = monto
    @abstractmethod
    def procesar_pago (self) ->None:
        pass

# crearemos la clase hijos :


class PagoBancario(SistemaPago, GenerarReporte):
    def procesar_pago (self) ->None:
        print(f"Procesando tranferencia bancaria por ${self.monto:,.2f}")
    def generar_reporte(self) ->None:
        print("Reporte de pago realizado mediante sistema bancario")

class PagoCripto(SistemaPago):
    def procesar_pago (self) ->None:
        print(f"Procesando pago con cripto moneda por ${self.monto:,.2f}")

def ejecutar_pago(pago: SistemaPago) -> None:
    pago.procesar_pago()


def main() -> None:
    print("************SISTEMA DE PAGOS****************** ")
    pago1 = PagoBancario(500000)
    pago2 = PagoCripto(7500000)
    ejecutar_pago(pago1)
    ejecutar_pago(pago2)


if __name__ == "__main__":
  main()


```

```python
#EJERCICIO ZOOLOGICO
#ENUNCIADO
#vamos a construir un modelo de zoologico donde existen animales genericos (clase abstacta)
# y tendremos animales especificos que hereden comportamientos e implenantas  la accion hacer_sonudo mediante  polimorfismo

from abc import ABC, abstractmethod
from typing import  Optional
class Movible(ABC):
   @abstractmethod
   def mover(self) -> None:
    pass
class Animal(ABC):
   def __init__(self , nombre: str) -> None:
     # 3 El atributo privado (Encapsulado )
      self.__nombre: str = ""
      self.nombre = nombre     # trampa para enmarcarar el setter para la asignacion
   @property  # decorador ,  hace getter
   def nombre(self) ->str:
     return self.__nombre
   @nombre.setter
   def nombre(self , valor: str) -> None: #aseguramos que el nombre venga de una cadena de texto valida
     if isinstance(valor, str) and valor.strip():
       self.__nombre = valor.strip().title() # asigna el nuevo nombre quita espacios en blanco y pone primera letra en mayuscula
     else:
       raise ValueError("El nombre debe ser una cadena de texto valida")
   @abstractmethod
   def sonido(self) -> None:
      pass
    # crearemos las clases hijas
class Perro(Animal):
   def sonido(self) -> None:
       print(f"{self.nombre}  (Perro) dice : guau")
class Gato(Animal):
   def sonido(self) -> None:
       print(f"{self.nombre}  (Gato) dice : miau")
class Vaca(Animal):
   def sonido(self) -> None:
       print(f"{self.nombre}  (Vaca) dice : Muuu")
class Leon(Animal , Movible):
   def sonido(self) -> None:
       print(f"{self.nombre}  (Leon) dice : Booooarr")
   def mover(self) -> None:
       print(f"{self.nombre} Camina sigilosamente por la sabana..")
#FUNCIONES POLIMORFICAS
def hacer_sonido(animal: Animal) -> None:
    animal.sonido()
def desplazar(animal: Movible) -> None:
    animal.mover()
  # CREAMOS MAIN

def main() -> None:
   try:
      animales =  [
          Perro("Rocky"),
          Gato("Misu"),
          Vaca("Lola"),
          Leon("Simba")
    ]
      print("==POLIMOSFISMO EN MODELO ZOOLOGICO === \n")
      for animal in animales:
        hacer_sonido(animal)
      print("\n MOVIMIENTO +++")
      for animal in animales:
        if isinstance(animal , Movible):
          desplazar(animal)
      print("Actualizacion de los nombres")
      animales[0].nombre ="Max"
      print(f"El nuevo nombre del parro es {animales[0].nombre}")
      for animal in animales:
        hacer_sonido(animal)
   except ValueError as e:
     print(f"Error de validacion {e}")
if __name__ == "__main__":
   main()














```
