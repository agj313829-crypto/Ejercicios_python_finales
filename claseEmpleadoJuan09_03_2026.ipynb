```python
class Empleado:
    def __init__(self,nombre: str):
        # validaremos el nombre atilizando metodos estaticos
       if not Empleado.validar_nombre(nombre):
          raise ValueError("El nombre debe ser un texto valido.")
       self.nombre = nombre # atributo publico
    def trabajar(self) -> None:
        print(f"{self.nombre} esta realizando tareas generales")
    @staticmethod
    def validar_nombre(nombre: str)-> bool:
      return isinstance (nombre,str) and nombre.strip() !=""
class Gerente(Empleado):
    def trabajar(self)-> None:
       print(f"{self.nombre} esta coordinando el equipo y gestionando proyectos")
class Desarrollador(Empleado):
    def trabajar(self) ->None:
       print(f"{self.nombre}esta escribiendo codigo en python y resolviendo problemas  de codigo")
def ejecutar_tarea(empleado: Empleado)-> None: #funcionplimorfica general donde sobre escribiremos todo
    empleado.trabajar()
def main()-> None:
    print("\nEs valido 'Eduardo' ? ->", Empleado.validar_nombre('Eduardo'))
    print("¿Es valido 'Eduardo' ' ? ->", Empleado.validar_nombre(''))
    gerente = Gerente("Sofia")
    desarrollador = Desarrollador("Jose")
    ejecutar_tarea(gerente)
    ejecutar_tarea(desarrollador)
if __name__ == "__main__":
   main()







```

```python
from abc import ABC, abstractmethod

# VAMOS A CREAR UNA CLASE ABSTACTA

class Figura(ABC):
  @abstractmethod
  def calcular_area(self): # metodo es obligatorio para todas las subclases
     pass


# creacion de una clase normal y se invoca la abc
class  Circulo(Figura):
  def __init__(self,radio):
    self.radio = radio
  def calcular_area(self):
      return  3.1416 * (self.radio ** 2)

  # crearemos la funcion que utlizara polimorfismo
def mostrar_area(figura: Figura):
   print(f"Area: {figura.calcular_area():,.2f}")

def main():
  circulo = Circulo (5) # creacion del objeto
  mostrar_area(circulo) # polimosfismo puro
if __name__ == "__main__":
   main()

```

```python
#Algoritmo metodo de pago
from abc import ABC, abstractmethod
class GenerarReporte(ABC):
  @abstractmethod
  def generar_reporte(self) ->None:
    pass
#**** FIN DE LA CLASE PADRE EL METODO SE REPETIRA EN TODAS LAS CLASES
class SistemaPago(ABC):
  def __init__(self,monto:float)->None:
    self.monto = monto
    @abstractmethod
    def procesar_pago(self)->None:
      pass

class PagoBancario(SistemaPago, GenerarReporte):
   def procesar_pago(self)->None:
    print(f"procesando transferencia bancaria por {self.monto:,.2f}")
   def generar_reporte(self)-> None:
    print(f"Reporte de pago bancario por {self.monto:,.2f}")

class PagoCriptomoneda(SistemaPago):
  def procesar_pago(self) -> None:
    print(f"Procesando pago con Criptomoneda por {self.monto:,.2f}")

def ejecutar_pago(pago: SistemaPago) ->None:
  pago.procesar_pago()


def main():
  pago1 = PagoBancario(50000)
  pago2 = PagoCriptomoneda(7600000)
  ejecutar_pago(pago1)
  pago1.generar_reporte()
if __name__ == "__main__":
  main()





```
