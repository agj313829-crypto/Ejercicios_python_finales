```python
#1. Agregar nuevos clientes append()
def agregar_cliente(lista_clientes , nombre):
  if isinstance(nombre, str) and 2 <= len (nombre) <= 50:
    lista_clientes.append(nombre.title())
    print(f"Cliente agregado: {nombre.title()}")
  else:
    print("Nombre de cliente invalido  debe tener entre 2 a 50 caracteres")
#2. Recorrer la lista y mostrar los clientes for
def mostrar_clientes(lista_clientes):
  for cliente in lista_clientes:
    print(f"-> {cliente}")
#3 Modificar un nombre en caso de error
def modificar_cliente(lista_clientes,indice, nuevo_nombre):
    if not isinstance(nuevo_nombre, str) and 2 <= len(nuevo_nombre) <= 50:
      print("Nombre de cliente invalido debe tener 2 a 50 caracteres")
      return # termina la funcion
    if 0 <= indice < len(lista_clientes):
        nombre_original = lista_clientes[indice]
        lista_clientes[indice] = nuevo_nombre.title()
        print(f"El cliente {nombre_original} fue reemplazado por: {nuevo_nombre.title()}")
    else:
      print("No se encontro el dato, numero de la lista erroneo")
#4.Borrar un cliente
def borrar_cliente(lista_clientes, indice):
    if 0 <= indice < len(lista_clientes):
      borrar = lista_clientes.pop(indice)
      print(f"cliente {borrar}, se borro de la lista de clientes.")
    else:
        print("No se encontro el cliente, cliente no borrado de la lista.")
def main():
# simulacion empresarial
   clientes = ["hugo" , "paco" , "luis" , "mynnie"]
   print("clientes actuales")
   mostrar_clientes(clientes )
   #1. Agregar nuevos clientes append()
   mostrar_clientes(clientes,)
   agregar_cliente(clientes, "alexander")
   mostrar_clientes(clientes)
   modificar_cliente(clientes,1,"diego")
   mostrar_clientes(clientes)
   borrar_cliente(clientes,3)
if __name__ == "__main__":
    main()



```
