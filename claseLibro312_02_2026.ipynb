```python
class Libro:
  def __init__(self, titulo, precio):
    self.__titulo = titulo
    self.__precio = precio
  @property  # DECORADOR  con este comando creo el getter (get) muestra la informacion de los atributos de cada clase
  def precio(self):
      return self.__precio

  @property
  def titulo(self):
    return self.__titulo

  @precio.setter  #modificar
  def precio(self, nuevo_precio):
    if isinstance(nuevo_precio,(int , float)) and nuevo_precio >0:
        self.__precio = nuevo_precio
    else:
        print("Debe ingresar un numero valido en el precio")

  @titulo.setter
  def titulo(self,nuevo_titulo):
     if isinstance(nuevo_titulo,str):
        self.__titulo =nuevo_titulo
     else:
      print("Debe ingresar un texto valido en el titulo")
  def mostrar_info(self):
      print(f"\nEl titulo del libro es: {self.__titulo}")
      print(f"El precio de el libro es: {self.__precio}\n")
      print(f"\nEl titulo del libro es: {self.__titulo},El precio del libro es: {self.__precio}\n")

def main():
    print("========CREACION DE OBJETOS EN POO=====\n")
    libro1 = Libro("la melancolia de los feos",40000)   # Creo mi primer objeto
    libro2 = Libro("leidy masacre", 20000)

    print("el nombre del libro 1 es:")
    print(libro1.titulo)


    print("El nombre de el libro  es: ", libro1.titulo, "Precio:", libro1.precio)

    print("todos los datos")
    libro1.mostrar_info()

    print("Actualizaremos el titulo\n")
    libro1.titulo = input("ingrese el nuevo titulo: ")
    print("Titulo actualizado: ahora el libro 1 se llama:" ,libro1.titulo)

    print("el precio del libro es:",libro1.precio," Actualizaremos este precio por otro \n")
    libro1.precio = input("ingrese el nuevo precio 1: ")
    print("\nPrecio actualizado: ahora el libro 1 se cuesta:" ,libro1.precio)
    print("\nlos nuevos datos del libro 1 son:")
    libro1.mostrar_info()



    print("el nombre del libro 2 es:")
    print(libro2.titulo)


    print("El nombre de el libro  es: ", libro2.titulo, "Precio:", libro2.precio)

    print("todos los datos")
    libro2.mostrar_info()


    print("Actualizaremos el titulo\n")
    libro2.titulo = input("ingrese el nuevo titulo: ")
    print("Titulo actualizado: ahora el libro 2 se llama:" ,libro1.titulo)


    print("el precio del libro es:",libro2.precio," Actualizaremos este precio por otro \n")
    libro2.precio = input("ingrese el nuevo precio 2: ")
    print("\nPrecio actualizado: ahora el libro 2 se cuesta:" ,libro2.precio)
    print("\nlos nuevos datos del libro 2 son:")
    libro2.mostrar_info()






if __name__ == "__main__":
    main()


```
